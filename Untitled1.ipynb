```python
from fastbook import *

# GUI components.
from fastai.vision.widgets import *
```

```python
# Image dataset uploaded into Kaggle.
path_dataset_root = Path('/content/grass')
path_image = Path('grassture/IMG_6363.JPG')

# Test open an image from the dataset.
im = Image.open(path_dataset_root / path_image)
im.to_thumb(128,128)
```

```python
# Get image files names for files recursively from the root folder.
image_file_names = get_image_files(path_dataset_root)
image_file_names
```

```python
# Data block with images from dataset, images labeled with parent folder names, example: 'GrassTrue', 'GrassFalse'.
# Data splitted in one learning and one validation part.
grass = DataBlock(
    blocks=(ImageBlock, CategoryBlock),
    get_items=get_image_files,
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=Resize(128))
```

```python
# Create a dataloaders object with image dataset set as source.
dls = grass.dataloaders(path_dataset_root)
```

```python
# Show some example of dataloader data.
dls.valid.show_batch(max_n=4, nrows=1)
```

```python
# Random transformation of images
# i.e. multiply the images sources from dataset into variants (rotation, brightness etc) for better learning.
grass = grass.new(
    item_tfms=RandomResizedCrop(224, min_scale=0.5),
    batch_tfms=aug_transforms())
dls = grass.dataloaders(path_dataset_root)
```

```python
# Learning and a number of fine-tuning epochs.
learn = vision_learner(dls, resnet18, metrics=error_rate)
learn.fine_tune(4)
```

```python
# Visualize the prediction results, predicted versus actual image.
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

```python
# Visualize top losses.
interp.plot_top_losses(5, nrows=5)
```

```python
# Show GUI for data cleaner tool.
cleaner = ImageClassifierCleaner(learn)
cleaner
```

```python
# Save a copy of the architecture and parameters.
learn.export()
```

```python
# List the saved file.
path = Path()
path.ls(file_exts='.pkl')
```

```python
# Load the saved architecture and parameters.
learn_inf = load_learner(path/'export.pkl')
```

```python
# Execute a prediction/inference.
learn_inf.predict(path_dataset_root / path_image)
```

```python
# List the categories to predict from dataloaders object.
learn_inf.dls.vocab
```

```python
# File upload GUI component.
btn_upload = widgets.FileUpload()
btn_upload
```

```python
# Fake an uploaded image to pass automatic code test during code saving operation.
btn_upload = SimpleNamespace(data = [path_dataset_root / path_image])
```

```python
# Show the uploaded image.
img = PILImage.create(btn_upload.data[-1])
img
```

```python
# Show the uploaded image as a thumbnail.
out_pl = widgets.Output()
out_pl.clear_output()
with out_pl: display(img.to_thumb(128,128))
out_pl
```

```python
# Predict if image is grass or not.
pred,pred_idx,probs = learn_inf.predict(img)
```

```python
# Display the prediction and probability.
lbl_pred = widgets.Label()
lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'
lbl_pred
```

```python
# Create a run GUI button 'Classify'.
btn_run = widgets.Button(description='Classify')
btn_run
```

```python
# On classify click event:
# get uploaded image, display image, predict image, display prediction and probability.
def on_click_classify(change):
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

btn_run.on_click(on_click_classify)
```
